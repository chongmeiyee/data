```python
import numpy as np
import pandas as pd
from linearmodels import PanelOLS
from linearmodels import FamaMacBeth
import statsmodels.formula.api as sm
```

```python
data = pd.read_csv('pdata')
data['date'] = pd.to_datetime(data['date'])
data.set_index(['TICKER', 'date'], inplace=True)
```

```python
mod = PanelOLS.from_formula("RETX~1+retail_ratio+ROA+LEV+turnover+TimeEffects",data=data)
```

```python
res = mod.fit(cov_type='clustered')
```

```python
print(res)
```

```python
res1 = mod.fit(cov_type='clustered',cluster_entity=True, cluster_time=True)
```

```python
print(res1)
```

```python
1 - (1-res.rsquared)*(len(data.RETX)-1)/(len(data.RETX)-4-1)
```
